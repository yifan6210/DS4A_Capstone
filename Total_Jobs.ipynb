```python
import pandas as pd
import numpy as np
import os
import seaborn as sn
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp
```

```python
df = pd.read_csv("merged.csv")
df['sex'] = df['sex'].replace(1, 'male')
df['sex'] = df['sex'].replace(2, 'female')
```

```python
sum_jobs = df.iloc[:, [0,40]].groupby('ID').sum() #is.na().sum() = 0, there does not exist NAs
sum_jobs = sum_jobs.reset_index().rename(columns={"N_Jobs": "T_Jobs"})
sum_jobs 
```

```python
year2016 = df[df['year'] == 2016]
year2016.isnull().sum()/year2016.shape[0]*100
year2016 = pd.merge(year2016,sum_jobs, on = 'ID')
year2016.groupby('ID').count() #12686 participants recorded
```
